```python
import numpy as np
import os

src_folder = '2024'
dst_folder = '2024_mean'
os.makedirs(dst_folder, exist_ok=True)

file_list = [f for f in os.listdir(src_folder) if f.endswith('.npy')]

for file in file_list:
    src_path = os.path.join(src_folder, file)
    dst_path = os.path.join(dst_folder, file)
    data = np.load(src_path)           # shape: (80, 768)
    col_mean = data.mean(axis=0)       # shape: (768,)
    # col_mean = col_mean[None, :]       # shape: (1, 768)
    np.save(dst_path, col_mean)
    print(f"{file}: origin shape {data.shape} -> mean shape {col_mean.shape} | saved to {dst_path}")
```

```python
import os
import numpy as np
import torch

def read_CKG_mean_emb(folder_path):
    all_emb = []
    for fname in sorted(os.listdir(folder_path)):
        if fname.endswith('.npy'):
            print(f"Loading {fname}...")
            emb = np.load(os.path.join(folder_path, fname))   # shape: (768,)
            all_emb.append(emb)
    all_emb = np.stack(all_emb, axis=0)   # shape: (文件数, 768)
    tensor_emb = torch.tensor(all_emb, dtype=torch.float32)
    print(f"Loaded mean embeddings: shape {tensor_emb.shape}")
    return tensor_emb

tensor_emb = read_CKG_mean_emb("2023_mean")
print(tensor_emb.shape)
# 保存
torch.save(tensor_emb, '2023_mean.pt')
```
